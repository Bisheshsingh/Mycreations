
# DROPPING STRING COLUMNS

```python
tr=df.drop("Monthly Debt",axis=1)
tr=tr.drop("Loan ID",axis=1)
tr=tr.drop("Customer ID",axis=1)
tr=tr.drop("Loan Status",axis=1)
tr=tr.drop("Term",axis=1)
tr=tr.drop("Years in current job",axis=1)
tr=tr.drop("Purpose",axis=1)
tr=tr.drop("Home Ownership",axis=1)
tr=tr.fillna(0)
tr=tr[:100000]
tr_lab=df['Monthly Debt'].copy()
tr.info()
```

```python
tr_lab=tr_lab[:100000]
tr_lab
```

# TRAINING

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")
imputer.fit(tr)
x=imputer.transform(tr)
```

```python
tr_x = pd.DataFrame(x, columns=tr.columns)
tr_x.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([('imputer', SimpleImputer(strategy="median")), ('std_scaler', StandardScaler()),])
tr_num=my_pipeline.fit_transform(tr_x)
sd=pd.DataFrame(tr_num)
sd.info()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
#model = LinearRegression()
#model = DecisionTreeRegressor()
model = RandomForestRegressor(n_estimators=20,random_state=0)
```

```python
model.fit(tr_num,tr_lab)
```

```python
sdata=tr[:5]
tr[:5]
```

```python
pdata=my_pipeline.transform(sdata)
pdata
```

# TESTING MODEL


```python
ted=pd.read_csv("credit_test.csv")
ted.head()
```

```python
te=ted.drop("Monthly Debt",axis=1)
te_lab=ted['Monthly Debt'].copy()
te=te.drop("Loan ID",axis=1)
te=te.drop("Customer ID",axis=1)
te=te.drop("Term",axis=1)
te=te.drop("Years in current job",axis=1)
te=te.drop("Purpose",axis=1)
te=te.drop("Home Ownership",axis=1)
te=te.fillna(0)
te=te[:10000]
te.head()
```

```python
te_lab=te_lab[:10000]
te_lab
```

```python
from sklearn.metrics import mean_squared_error
trp = model.predict(my_pipeline.transform(te))
mse = mean_squared_error(te_lab,trp)
rmse = np.sqrt(mse)
rmse
```
